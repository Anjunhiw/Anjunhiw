```python
""" 생성형 인공지능 원리
토큰화와 임베딩
1. 토큰화
1) 컴퓨터는 숫자로 된 데이터를 처리하는 것이 가장 익숙하다.
2) 하지만 사람이 쓰는 자연어는 숫자가 아니라 단어와 문장으로 구성되어 있다.
3) 그렇기 때문에 인공지능이 사람의 언어를 이해하려면 문장을 작은 조각(토큰)으로 나누고 이를 숫자로 변환해야 한다.
4) 이렇듯 문장을 나누는 과정을 토큰화(토크나이제이션)라고 한다.

예시

1) 음절단위 : 문장을 한글자 씩 나누는 규칙
입력 : "아버지가 방에 들어가신다"
출력 : ["아", "버", "지", "가", "방", "에", "들", "어", "가", "신", "다"]

2) 서브워드 : 단어보다 작은 단위(서브워드)로 나누어 자연어 처리모델이 더 효과적으로 학습, 기존에 있던 단어도 효과적으로 처리 가능
입력 : "초해상도영상처리"
출력 : ["초", "해상도", "영상", "처리"]

"""
```

```python
# 토큰화 실습
from transformers import AutoTokenizer # 허킹페이스에서 제공하는 강력한 NLP 라이브러리중 트랜스포머 라이브러리 사용, 특정모델에 맞는 토크나이저를 자동으로 불러오는 클래스인 AutoTokenizer 사용
model_id = "microsoft/phi-4" # 사용할 인공지능 모델 : 마이크로소프트에서 개발한 phi-4(다른 모델에 비해 경량화, 한글 지원 가능)
tokenizer = AutoTokenizer.from_pretrained(model_id) # AutoTokenizer.from_pretrained(model_id)를 실행하면 model_id에 해당하는 모델의 토크나이저를 자동으로 내려받음
```

```python
# 토큰화 테스트
text = "I am studying tokenization." # 테스트할 문장 저장
tokens = tokenizer.tokenize(text) # 문장을 여러개의 토큰(단어 단위)로 분해
print(tokens) # 분해된 토큰들 출력

# 'I' : 단어 자체가 짧고 일반적이라 그대로 유지
# 'Ġam' : Ġ문자는 특수토큰으로 공백을 의미, 원래 문장에서 'am(공백 + am)' 형태
# 'Ġstudying' : 'studying(공백 + studying)' 형태
# 'Ġtoken' : token(공백 + token)형태
# 'ization' : 'tokenization'라는 단어가 데이터에서 자중 등장하는 'token'과 'ization'으로 분할되었기 때문
# '.' : 마침표도 별도의 토큰으로 인식
```

```python
# 토큰화된 문장을 숫자로 변환 : 토큰 ID 변환(token ID mapping)
# 각 토큰은 사용하는 언어 모델에 따라 고유한 숫자(token ID)를 갖고 있다.

token_ids = tokenizer.convert_tokens_to_ids(tokens)
print(token_ids)
```
